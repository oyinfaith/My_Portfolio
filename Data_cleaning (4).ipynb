# Checking LongName to remove special character in it or irregularity in it.

```python
df[['LongName']].head(15)
```

```python
#importing libraries that helps to check special characters in a column
import string
import re
alphabet = string.ascii_letters+string.punctuation
```

```python
#To check if there are special characters in Name and LongName column of the players
df.LongName.str.strip(alphabet).astype(bool).any()
```

```python
#Removing the special character in LongName using a function.
def strip_character(dataCol):
    r = re.compile(r'[^a-zA-Z !@#$%&*_+-=|\:";<>,./()[\]{}\']')
    return r.sub('', dataCol)

df['LongName'] = df['LongName'].apply(strip_character)
```

# Converting inches to cm for Height

```python
df.Height.unique()
```

```python
def inches_to_cm(value):
    if isinstance(value, str) and value[-2:]=='cm':
        value = value.strip('cm')
        return value
    else:
        return value
```

```python
df['Height'] = df['Height'].apply(inches_to_cm)
```

```python
df['Height'].unique()
```

```python
def sign(value):
    if isinstance (value, str) and value[-1]=='"':
        value = value.replace("'",'')
        value = value.replace('"', '')
        value = value.replace('\'','')
        value = float((int(value[0]))*12 + int(value[-1])*2.54)
        return value
    else:
        return float(value)
    
```

```python
df['Height_cm'] = df['Height'].apply(sign)
```

```python
df['Height'].unique()
```

```python
df['Height_cm'] = df.Height_cm.astype(float)
df.drop('Height', axis=1, inplace = True)
```

```python
df.Height_cm.unique()
```

# Working on Weight to convert pounds to kg

```python
df['Weight'].unique()
```

```python
def remove_k(val):
    if isinstance(val, str) and val[-2:] =='kg':
        val = val.strip('kg')
        return val
    else:
        return val
```

```python
df['Weight'] = df['Weight'].apply(remove_k)
```

```python
df['Weight']
```

```python
def convert_pounds_kg(val):
    if isinstance(val, str) and val[-3:] =='lbs':
        val = str(round(float(val.replace('lbs', ''))*0.454))
        return val
    else:
        return val
```

```python
df['Weight_kg'] = df['Weight'].apply(convert_pounds_kg)
df['Weight_kg'] = df.Weight_kg.astype(float)
```

```python
df['Weight'].unique()
```

```python
df.drop('Weight', axis=1, inplace =True)
df['Weight_kg'].unique()
```

# Working on the Value

```python
df['Value'].unique()
df['Wage'].unique()
df['Release Clause'].unique()
```

```python
df['Value_Euro'] = df['Value'].str.strip('€')
df['Release Clause_Euro'] = df['Release Clause'].str.strip('€')
df['Wage_Euro'] = df['Wage'].str.strip('€')
```

```python
def convert_to_dollar(dollar):
    if 'M' in dollar:
        dollar = dollar.strip('M')
        millon = float(dollar) * 1000000
        return millon
    else:
        dollar = dollar.strip('K')
        thousand = float(dollar) * 1000
        return thousand
```

```python
df['Value_Euro'] = df['Value_Euro'].apply(lambda X: convert_to_dollar(X))
df['Release Clause_Euro'] = df['Release Clause_Euro'].apply(lambda X: convert_to_dollar(X))
df['Wage_Euro'] = df['Wage_Euro'].apply(lambda X: convert_to_dollar(X))
```
